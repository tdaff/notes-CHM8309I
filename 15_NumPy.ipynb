NumPy and SciPy
===============

Python has iterables and containers as built in types, but not a so-called **array**.

+ `list`s are **mutable sequences** that can contain any other object and can be modified in-place. 
+ `tuple`s are **immutable sequences** that cannot be modified once created, but offer some runtime optimisations.

We have already seen how we can treat these objects more array like, but they don't behave as we might expect mathematically.

```python
array_like = [[1.2, 2.2], [-4.3, 0.2]]
```

```python
print(array_like[0])  # Index items
print(array_like[1][1])  # Index nested items
```

```python
print(array_like*2)  # Doesn't do what we expect numerically
print([[x*2 for x in row] for row in array_like])  # Map and filter
```

Getting mathematical behaviour
------------------------------

To get the behaviour that we want, we can iteratre over our arrarys with `for` loops (take care with nested lists), or even subclass and construct our own array-like that behaves as we want.

```python
from operator import add, mul
from itertools import chain

class NumList(list):
    """Subclass list to make it more vectorish."""
    def __mul__(self, value):
        """Elementwise multiplication."""
        #Remeber that the opertion must return a NumList not just a list
        return NumList([mul(x, value) for x in self])
    def __add__(self, value):
        """Elementwise add for equal length or add value to each element."""
        if isinstance(value, list):
            return NumList([add(x, y) for x, y in zip(self, value)])
        else:
            return NumList([add(x, value) for x in self])
```

```python
my_num_list = NumList([12.3, 4.4, 0.9])
```

```python
print(my_num_list*5)
print(my_num_list+my_num_list)
print(my_num_list+3)
```

`dtype`
-------

`numpy.ndarray` types differ from lists in other ways besides their mathematical behaviour:

+ Python lists can contain any mixture of object and are dynamically typed.
    + Implementating maths for lists is inefficient because of the dynamic typing.
+ NumPy arrays are homogeneous and statically typed -- the `dtype`.
    + With knowledge of the types, fast maths can be implemented in a compiled language (C and Fortran is used).
+ Numpy arrays are memory efficient.

```python
my_array.dtype  # a property of the whole array
```

```python
my_array.astype('int')  # convert dtypes
```

```python
np.array([2, 3, 4], dtype='float')  # becomes 'float64'
```

As the `ones` example shows, unlike the built in sequences, ndarrays can have a dimensionality that can be changed.

Slicing and assignment
----------------------

_Slices_ can be used for views and assignment. There are also ways to do _fancy_ slicing in different directions.

```python
my_other_array[1, 1]  # index an item
```

```python
my_other_array[2, 2] = -10
```

```python
my_other_array[:, 2]  # column 2
```

```python
# slices can be start:stop:stride on each axis
my_other_array[0:2, ::2] = -1  # assign to slices
my_other_array
```

```python
mask = (my_other_array > 0)  # True/False for positive/negative
mask
```

```python
my_other_array[mask]  # take all items that are True in mask
```

Vector maths
------------

You can also carry out some vector math type operations between arrays:

```python
one, two = np.array([1.2, 2.2, 3.3]), np.array([-0.4, 3.2, 1.8])

# CAREFUL: elementwise multiplication
one*two
```

```python
# probably want the dot product
np.dot(one, two)
```

```python
np.cross(one, two)
```

```python
# matrix is an array that does matrix algebra with operators
matrix_one, matrix_two = np.matrix(one), np.matrix(two)
```

```python
matrix_one*matrix_two.T  # .T is the transpose
```

Vectorised functions
--------------------

Some examples already showed elementwise _broadcast_ operations. NumPy also comes with a number of functions that are optimised to work on all elements in an array at once. This is one of the reasons it is so fast. You can see the long complete list of what numpy can do at [http://www.scipy.org/Numpy_Example_List_With_Doc](http://www.scipy.org/Numpy_Example_List_With_Doc) where they include examples for everything.

The examples are generated from the docstrings, so you can get all this information at any time by checking the docstrings (use `help(np.function)`) and the notebook or a decent IDE will pop up everything you need to know, when you need it.

```python
np.square(three_by_three)  # broadcast to every element
```

Functions that are just combinations of vectorised operators also broadcast on arrays. If that doesen't work (for example, some conditional expressions don't work) we can get it to vectorize.

```python
def fractional_in_cell(coordinate):
    """Return the in-cell part of the fractional coordinate."""
    if coordinate > 1.0 or coordinate < 0.0:  # breaks vectorisation
        return coordinate - np.floor(coordinate)
    else:
        return coordinate

#fractional_in_cell(three_by_three)  # gives a ValueError
# The modulus version works fine
```

```python
# Can do automatic vectorisation in many cases
fractional_in_cell_v = np.vectorize(fractional_in_cell)
```

```python
fractional_in_cell_v(three_by_three)
```

`ndimage`
---------

```python
%matplotlib inline
from matplotlib.pyplot import imshow

noisy_grid_data = np.random.random((50, 200))
_ = imshow(noisy_grid_data)
```

```python
import scipy
# submodules need to be imported separately
import scipy.ndimage

# ndimage provides image processing type functions
filtered = scipy.ndimage.gaussian_filter(noisy_grid_data, 2)
_ = imshow(scipy.ndimage.interpolation.zoom(filtered,  [1.2, 0.75]))
```

`spatial`
---------

```python
import scipy.spatial

# maybe a set of atoms positions?
my_particles = np.array([[0.0, 0.0, 0.0], [2.3, 7.4, 1.1], [1.4, -1.1, 0.2]])

# spatial can give very fast distance calculations between arrays of points
scipy.spatial.distance.cdist(my_particles, my_particles)
```
