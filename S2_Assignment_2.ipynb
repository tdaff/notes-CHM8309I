```python
"""
Assignment 2

The idea was to get you to think about design and producing
object oreinted code that integrates with derivative works.
Because of this, I have not explicitly defined units or used
a complicated dynamics integrator as the interface is more
important than the details of the implementation in this case.

"""

class DynamicsMixin(object):
    """Provide movement methods for an object that exposes weight."""
    # The mixin provides all the members to make something move
    # except for weight which must be provided by any child
    # class so that it can have acceleration

    def __init__(self):
        """Add movement methods and attributes for a simple integrator."""
        # Assume that these three are enough to describe motion
        self.position = [0.0, 0.0, 0.0]
        self.velocity = [0.0, 0.0, 0.0]
        self.force = [0.0, 0.0, 0.0]
        self.fixed = False
        # Keep the time internal as a tracker of state
        self._time = 0
        # Don't expose history as the implementation is not
        # decided yet
        self._historical_position = {}

    def timestep(self, delta_time=1.0):
        """
        Very basic dynamics integrator assumes constant
        velocity and acceleration over the duration of delta_time.
        """
        # in a real MD code, this would be a real integrator
        # like a Verlet one with specific units
        if not self.fixed:
            # rudimentary fixing of position should be better
            self.position = vector_add(self.position,
                                       vector_scale(self.velocity, delta_time))
        self.velocity = vector_add(self.velocity,
                                   vector_scale(self.acceleration, delta_time))
        self._time += delta_time

    def scale_velocity_to_temperature(self, temperature):
        """Use the given temperature to rescale the velocity."""
        # kinetic energy has a square relationship with temperature
        # so scale by the sqaure root of the ratios
        scale_factor = (temperature/self.temperature)**0.5
        self.velocity = vector_scale(self.velocity, scale_factor)
        
    def history_str(self, steps=None):
        """
        Return a formatted output of the position history for the previous number
        of steps where the format is 'time <position>'

        """
        # actual history not exposed for this version
        # TODO(tdaff): expose numeric history in a nice format for v1 API
        if steps is None:
            steps = len(self._historical_position)
        return "\n".join("{:<6} ".format(time) +
                         "<{}>".format(self._historical_position[time])
                         for time in sorted(self._historical_position)[-steps:])
        
    def _get_position(self):
        """Getter for the position."""
        # the actual position is private and encapsulated
        # so that the implementation details with the
        # automatic history are hidden from the interface
        return self._position

    def _set_position(self, value):
        """
        Set the position to a new value and store the previous
        one in the history. Change position before changing time
        as the current time is recorded.
        """
        if hasattr(self, '_position'):
            # we only store the position in the history if this is
            # not the first step. Not perfect.
            self._historical_position[self._time] = self._position
            self._position = value
        else:
            self._position = value

    # having position as a property helps to neatly
    # hide the history tracking
    position = property(_get_position, _set_position)

    @property
    def acceleration(self):
        """Acceleration vector for moving object derived from the force."""
        # the property decorator makes this a read only attribute
        return [x/self.weight for x in self.force]

    @property
    def speed(self):
        """Scalar speed from velocity."""
        # don't want to be able to set the speed as it
        # has no direction information
        return pow(sum(x**2 for x in self.velocity), 0.5)

    @property
    def temperature(self):
        """Effective atom temperature."""
        # based on kinetic energy = 1/2mv^2 = 3/2kT
        # depending on our units there would be a boltzman
        # scaling in there
        return (self.weight * self.speed ** 2) / 3


class Atom(object):
    """Chemical atom."""
    # This is a very basic Atom that just has some chemical
    # properties and mostly contains data
    def __init__(self, weight=0, name=None, element=None, charge=None, params=None):
        self.name = name
        self.element = element
        self.weight = weight
        self.charge = charge
        self.force_field_parameters = params

    def __str__(self):
        """Return a nice string representation of the object."""
        # override the standard operation for print so that
        # the atom has a more descriptive look
        return "Atom: {self.name} ({self.element})".format(self=self)


class DynamicAtom(Atom, DynamicsMixin):
    """Chemical atom for a dynamics simulation."""
    def __init__(self, weight=0, name=None, element=None, charge=None, params=None):
        # since we are inheriting from two classes we have
        # to initialise the members in the correct way from
        # their initialisation methods
        Atom.__init__(self, weight, name, element, charge, params)
        DynamicsMixin.__init__(self)

    def __str__(self):
        """Return a nice string representation of the object."""
        # overriding this method from the parent class makes
        # the object nicely polymorphic so that it shows extra
        # information
        return ("Atom: {self.name} ({self.element})\n"
                "pos <{self.position}>\nvel <{self.velocity}>").format(self=self)


def vector_scale(vector, scale_factor):
    """Return the vector with elementwise scaling."""
    return [x*scale_factor for x in vector]


def vector_add(vector_one, vector_two):
    """Return elementwise sum of two vectors."""
    return [x + y for x, y in zip(vector_one, vector_two)]
```

assignment_atom
---------------

The assignment_atom module defines three classes with methods that can be used as chemical atoms, or moving objects for dynamics simulation. The `Atom` and `DynamicAtom` classes will work on thier own, but the `DynamicsMixin` should be inherited by an object with weight.

* `Atom` class is a static container for chemical information
* All keyword arguments are optional `weight=0, name=None, element=None, charge=None, params=None`
* it exposes the following attributes:
    + `name` (str) some identifier for the atom
    + `element` (str) chemical elemnt of the atom
    + `weight` (float) weight of this particular atom
    + `charge` (flaot) charge in atomic units
    + `force_field_parameters` (tuple) any parameters associated with the current force field; must be set extrnally.
* no methods are exposed, although `print()` will invoke an informational representation

--

* `DynamicsMixin` provides methods for making an `object` move it is expected that this is inherited by an object that exposes a weight attribute otherwise methods will fail.
* The class takes no keyword arguments and all attributes must be set manually
* The following attributes are exposed (all are initialised at 0.0 for a 3d system, but dynamics should work in any number of dimensions):
    + `position` (list) current location
    + `velocity` (list) current velocity with direction
    + `force` (list) external force experienced by object
    + `fixed` (bool) set True to stop position changing
* The following read only attributes are also accessible:
    + `acceleration` derived from the force (may be writeable in future)
    + `speed` scalar magnitude of current velocity
    + `temperature` effective atom temperature from kinetic energy
* The following methods are provided:
    + `timestep(self, delta_time=1.0)` move the object through time upadating the position and velocity according to the acceleration (very basic integrator)
    + `scale_velocity_to_temperature(self, temperature)` rescale the velocity based on the given temperature
    + `history_str(self, steps=None)` create a string representation of the previous `steps` number of timesteps.

--

* `DynamicAtom` inherits from both `Atom` and `DynamicsMixin` to provide an atom that can be used in dynamic simulation.
* All keyword arguments are optional `weight=0, name=None, element=None, charge=None, params=None` but weight should be greater than zero or weird things will happen.
* It exposes no further functionality other than the methods and attributes inherited from the parent classes.
